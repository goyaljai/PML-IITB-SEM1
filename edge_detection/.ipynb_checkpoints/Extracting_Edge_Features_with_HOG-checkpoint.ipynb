# Extracting Edge and HOG Features for IPL Grid Cells

This notebook populates the existing cell-level CSV with edge-based image features and HOG-style shape features.

The goal here is **not** to create train/test splits or train a model. The goals are:

1. Read the existing feature CSV.
2. Find each referenced image in the Hugging Face dataset.
3. Compute edge maps for each image.
4. Extract edge features for each cell using `x_min`, `y_min`, `x_max`, and `y_max`.
5. Extract HOG features for each cell to capture the direction and structure of gradients/edges.
6. Save a new populated CSV.

The output remains one row per `(image_file_name, cell_id)`.


## 1. Imports

These libraries are used for data handling, image loading, edge detection, HOG feature extraction, and downloading the image dataset.

The HOG features in this notebook are implemented using NumPy and OpenCV, so no extra `scikit-image` dependency is required.


````python
# Run this only if the libraries are not installed in your environment.
# %pip install pandas numpy opencv-python huggingface_hub

from pathlib import Path
from collections import defaultdict

import cv2
import numpy as np
import pandas as pd
from huggingface_hub import snapshot_download
````

## 2. Configuration

Change only these values if your file names or output location changes.

`TARGET_CSV` is the CSV that already contains rows for image cells. This notebook will populate its edge-feature columns and write a new CSV.


````python
# CSV that already contains one row per image-cell.
TARGET_CSV = "Edge_features.csv"

# Output CSV after edge + HOG features are populated.
OUTPUT_CSV = "Edge_features_populated_with_hog.csv"

# Hugging Face dataset containing the images.
HF_REPO_ID = "goyaljai/IPL-Player-Detection-IITB-PML"
HF_REPO_TYPE = "dataset"

# Project-required image size.
IMAGE_WIDTH = 800
IMAGE_HEIGHT = 600

# Project grid size.
GRID_ROWS = 8
GRID_COLS = 8

CELL_WIDTH = IMAGE_WIDTH // GRID_COLS      # 100 pixels
CELL_HEIGHT = IMAGE_HEIGHT // GRID_ROWS    # 75 pixels

# HOG settings.
# Each 100x75 cell is split into 2x2 smaller regions.
# For each region, we compute a 9-bin orientation histogram.
# Total HOG features per cell = 2 * 2 * 9 = 36.
HOG_PATCH_ROWS = 2
HOG_PATCH_COLS = 2
HOG_BINS = 9
HOG_FEATURE_COLUMNS = [
    f"f_hog_{i:03d}"
    for i in range(HOG_PATCH_ROWS * HOG_PATCH_COLS * HOG_BINS)
]

# Edge feature columns to populate.
# All model feature columns start with f_.
BASIC_EDGE_FEATURE_COLUMNS = [
    "f_edge_density",
    "f_canny_count",
    "f_sobel_mean",
    "f_sobel_std",
    "f_laplacian_var",
    "f_contour_count",
]

EDGE_FEATURE_COLUMNS = BASIC_EDGE_FEATURE_COLUMNS + HOG_FEATURE_COLUMNS

print("Number of basic edge features:", len(BASIC_EDGE_FEATURE_COLUMNS))
print("Number of HOG features:", len(HOG_FEATURE_COLUMNS))
print("Total edge/HOG feature columns:", len(EDGE_FEATURE_COLUMNS))
````

## 3. Read the target CSV

This notebook assumes the CSV already has the proposed cell-level structure.

Important columns used for feature extraction:

- `image_file_name`: image to load
- `cell_id` / `cell_num`: cell identifier
- `x_min`, `y_min`, `x_max`, and `y_max`: cell boundaries in the 800×600 image

The label columns such as `y_team_id` are not used to calculate edge or HOG features.


````python
df = pd.read_csv(TARGET_CSV)

print("Loaded target CSV")
print("Shape:", df.shape)
print("Columns:")
print(df.columns.tolist())

df.head()
````

## 4. Ensure required columns exist

The edge and HOG feature columns are created if they are missing. Existing values are kept unless they are overwritten during feature extraction.

If cell boundary columns are missing but `cell_num` is present, the notebook can recreate the 8×8 grid boundaries.


````python
# Add edge feature columns if absent.
for col in EDGE_FEATURE_COLUMNS:
    if col not in df.columns:
        df[col] = np.nan

# If cell boundaries are missing, recreate them from cell_num.
required_boundary_cols = ["x_min", "y_min", "x_max", "y_max"]
missing_boundary_cols = [col for col in required_boundary_cols if col not in df.columns]

if missing_boundary_cols:
    if "cell_num" not in df.columns:
        raise ValueError(
            "Missing boundary columns and cell_num is also missing. "
            "Need either x_min/y_min/x_max/y_max or cell_num."
        )

    print("Boundary columns missing. Recreating boundaries from cell_num...")

    cell_num_zero_based = df["cell_num"].astype(int) - 1
    df["grid_row"] = cell_num_zero_based // GRID_COLS
    df["grid_col"] = cell_num_zero_based % GRID_COLS

    df["x_min"] = df["grid_col"] * CELL_WIDTH
    df["y_min"] = df["grid_row"] * CELL_HEIGHT
    df["x_max"] = df["x_min"] + CELL_WIDTH
    df["y_max"] = df["y_min"] + CELL_HEIGHT

# Ensure boundary columns are numeric integers.
for col in required_boundary_cols:
    df[col] = df[col].astype(int)

print("Required edge columns and cell boundary columns are ready.")
df.head()
````

## 5. Download the image dataset and build an image index

This step downloads the Hugging Face dataset and builds a simple lookup table:

```text
image file name → full image path
```

We are **not** creating or analyzing train/test splits here. We are only locating the image files needed to populate edge features.

The dataset may internally store files in folders such as `train/` and `test/`, but this notebook treats all images as image files to be processed.


````python
dataset_dir = Path(
    snapshot_download(
        repo_id=HF_REPO_ID,
        repo_type=HF_REPO_TYPE
    )
)

print("Dataset directory:", dataset_dir)

# Build one image index across the entire dataset directory.
# This avoids split-specific logic because feature extraction only needs image files.
all_image_paths = (
    list(dataset_dir.rglob("*.jpg"))
    + list(dataset_dir.rglob("*.jpeg"))
    + list(dataset_dir.rglob("*.png"))
)

image_paths_by_name = defaultdict(list)
for path in all_image_paths:
    image_paths_by_name[path.name].append(path)

print("Total image files found:", len(all_image_paths))
print("Unique image file names found:", len(image_paths_by_name))

# Show duplicate file names, if any. Usually this should be empty.
duplicate_names = {name: paths for name, paths in image_paths_by_name.items() if len(paths) > 1}
print("Duplicate image file names:", len(duplicate_names))

if duplicate_names:
    first_name = next(iter(duplicate_names))
    print("Example duplicate name:", first_name)
    print(duplicate_names[first_name])
````

## 6. Helper function to locate an image

This function receives the image file name from the CSV and returns the path of the matching image.

If the same file name exists in more than one folder, it uses the first one and prints duplicate information in the previous cell. If duplicates become a real issue, we can add logic using `original_split` later, but that is not needed for simply creating edge features in most cases.


````python
def find_image_path(image_file_name):
    """Return the local image path for a file name from the CSV."""

    matches = image_paths_by_name.get(str(image_file_name), [])

    if not matches:
        return None

    return matches[0]
````

## 7. Compute edge and gradient maps for one full image

For each image, edge and gradient maps are computed once on the full 800×600 image.

Then cell-wise features are extracted by cropping the full-image maps. This is better than running Canny/Sobel separately on each small cell because it avoids boundary artifacts and is more efficient.

The HOG features use gradient magnitude and gradient orientation from the full image.


````python
def compute_edge_maps(image_bgr):
    """
    Compute edge-related maps for the full image.

    Input:
        image_bgr: OpenCV image in BGR format, already resized to 800x600.

    Output:
        Dictionary of full-image edge/gradient maps.
    """

    # Convert to grayscale because Canny, Sobel, Laplacian, and HOG are intensity-based.
    gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

    # Reduce small noise before edge detection.
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Binary edge map. Non-zero pixels represent detected edges.
    canny = cv2.Canny(blurred, threshold1=50, threshold2=150)

    # Sobel gradients capture intensity changes in x and y directions.
    sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

    # Gradient magnitude gives overall edge strength.
    sobel_mag = np.sqrt(sobel_x ** 2 + sobel_y ** 2)

    # Gradient orientation in degrees, folded into 0-180 degrees for unsigned HOG.
    sobel_angle = np.rad2deg(np.arctan2(sobel_y, sobel_x)) % 180

    # Laplacian captures high-frequency detail / sharpness.
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

    return {
        "canny": canny,
        "sobel_mag": sobel_mag,
        "sobel_angle": sobel_angle,
        "laplacian": laplacian,
    }
````

## 8. Extract edge and HOG features for one cell

Each row in the CSV represents one cell. This function crops the cell area from the full-image maps and calculates the required `f_` columns.

HOG features are calculated by splitting each grid cell into `2 × 2` patches and computing a `9`-bin orientation histogram in each patch. This gives `36` HOG columns per cell: `f_hog_000` to `f_hog_035`.


````python
def extract_hog_features(sobel_mag_cell, sobel_angle_cell):
    """
    Extract simple HOG-style features from one image cell.

    The cell is divided into HOG_PATCH_ROWS x HOG_PATCH_COLS patches.
    For each patch, we compute a histogram of gradient orientations weighted
    by gradient magnitude. The histograms are normalized so the values are
    comparable across cells.
    """

    hog_values = []

    # Split rows and columns into approximately equal patch boundaries.
    row_edges = np.linspace(0, sobel_mag_cell.shape[0], HOG_PATCH_ROWS + 1, dtype=int)
    col_edges = np.linspace(0, sobel_mag_cell.shape[1], HOG_PATCH_COLS + 1, dtype=int)

    for r in range(HOG_PATCH_ROWS):
        for c in range(HOG_PATCH_COLS):
            r0, r1 = row_edges[r], row_edges[r + 1]
            c0, c1 = col_edges[c], col_edges[c + 1]

            mag_patch = sobel_mag_cell[r0:r1, c0:c1].ravel()
            angle_patch = sobel_angle_cell[r0:r1, c0:c1].ravel()

            hist, _ = np.histogram(
                angle_patch,
                bins=HOG_BINS,
                range=(0, 180),
                weights=mag_patch
            )

            # Normalize each patch histogram to reduce scale differences.
            hist_sum = hist.sum()
            if hist_sum > 0:
                hist = hist / hist_sum

            hog_values.extend(hist.astype(float).tolist())

    return {
        col: float(value)
        for col, value in zip(HOG_FEATURE_COLUMNS, hog_values)
    }


def extract_edge_features(edge_maps, x_min, y_min, x_max, y_max):
    """
    Extract edge and HOG features for a single cell.

    The coordinates come directly from the CSV columns:
    x_min, y_min, x_max, y_max.
    """

    # Crop this cell from each full-image edge/gradient map.
    canny_cell = edge_maps["canny"][y_min:y_max, x_min:x_max]
    sobel_cell = edge_maps["sobel_mag"][y_min:y_max, x_min:x_max]
    sobel_angle_cell = edge_maps["sobel_angle"][y_min:y_max, x_min:x_max]
    laplacian_cell = edge_maps["laplacian"][y_min:y_max, x_min:x_max]

    cell_area = canny_cell.shape[0] * canny_cell.shape[1]

    # Count non-zero Canny pixels. These are edge pixels.
    canny_count = int(np.count_nonzero(canny_cell))

    # Normalized edge density. For a 100x75 cell, area is 7500.
    edge_density = canny_count / cell_area if cell_area > 0 else 0.0

    # Average and variability of edge strength.
    sobel_mean = float(np.mean(sobel_cell))
    sobel_std = float(np.std(sobel_cell))

    # Variance of Laplacian: rough sharpness / high-frequency detail measure.
    laplacian_var = float(np.var(laplacian_cell))

    # Contours from Canny map. More contours often means more local structure.
    contours, _ = cv2.findContours(
        canny_cell,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )
    contour_count = int(len(contours))

    features = {
        "f_edge_density": edge_density,
        "f_canny_count": canny_count,
        "f_sobel_mean": sobel_mean,
        "f_sobel_std": sobel_std,
        "f_laplacian_var": laplacian_var,
        "f_contour_count": contour_count,
    }

    # Add HOG features.
    features.update(
        extract_hog_features(
            sobel_mag_cell=sobel_cell,
            sobel_angle_cell=sobel_angle_cell
        )
    )

    return features
````

## 9. Populate the edge and HOG feature columns

This is the main processing loop.

The loop groups the CSV by `image_file_name` so that each image is read and processed only once. For each image, it computes the full edge/gradient maps once, then fills the edge and HOG columns for every cell row belonging to that image.


````python
missing_images = []
unreadable_images = []
processed_images = 0
processed_rows = 0

# Process one image at a time.
for image_file_name, group in df.groupby("image_file_name", sort=False):
    image_path = find_image_path(image_file_name)

    if image_path is None:
        missing_images.append(image_file_name)
        continue

    image_bgr = cv2.imread(str(image_path))

    if image_bgr is None:
        unreadable_images.append(str(image_path))
        continue

    # Ensure the image is exactly 800x600, matching the project grid.
    image_bgr = cv2.resize(image_bgr, (IMAGE_WIDTH, IMAGE_HEIGHT))

    # Compute edge maps once for this image.
    edge_maps = compute_edge_maps(image_bgr)

    # Populate edge features for each cell row of this image.
    for idx in group.index:
        x_min = int(df.at[idx, "x_min"])
        y_min = int(df.at[idx, "y_min"])
        x_max = int(df.at[idx, "x_max"])
        y_max = int(df.at[idx, "y_max"])

        features = extract_edge_features(
            edge_maps=edge_maps,
            x_min=x_min,
            y_min=y_min,
            x_max=x_max,
            y_max=y_max,
        )

        for col, value in features.items():
            df.at[idx, col] = value

        processed_rows += 1

    processed_images += 1

print("Processed images:", processed_images)
print("Processed cell rows:", processed_rows)
print("Missing images:", len(missing_images))
print("Unreadable images:", len(unreadable_images))

if missing_images:
    print("First missing images:", missing_images[:10])

if unreadable_images:
    print("First unreadable image paths:", unreadable_images[:10])
````

## 10. Quick validation

This cell checks whether the edge and HOG columns were populated. If any values are still missing, they are likely from images that were not found or could not be read.


````python
print("Missing values in edge/HOG columns:")
print(df[EDGE_FEATURE_COLUMNS].isna().sum())

print("Preview of populated edge/HOG columns:")
preview_feature_cols = BASIC_EDGE_FEATURE_COLUMNS + HOG_FEATURE_COLUMNS[:12]

df[[
    "image_file_name",
    "cell_id",
    "x_min",
    "y_min",
    "x_max",
    "y_max",
    *preview_feature_cols,
]].head(20)
````

## 11. Save the populated CSV

The original CSV is not overwritten. The populated result is saved to `OUTPUT_CSV`, which is now `Edge_features_populated_with_hog.csv`.


````python
df.to_csv(OUTPUT_CSV, index=False)

print("Saved populated CSV to:", OUTPUT_CSV)
print("Final shape:", df.shape)
````
